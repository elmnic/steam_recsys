```python
games_genres = pd.read_csv('../data/raw/csv/Games_Genres.csv')
games_genres.head()
```

# OneHotEncoding genres
Join all rows together by first changing the numbers to strings, then replacing '0' with '', group by 'appid' and finally join them in an aggregation

```python
# Turn into categorical vectors and combine all genres for each appid into one row
games_cat = pd.Categorical(games_genres['Genre'])
cat_dum = pd.get_dummies(games_cat, prefix='Genre').astype(str)
cat_dum = pd.concat([games_genres, cat_dum], axis=1).drop('Genre', axis=1)
cat_dum.head()
```

```python
maybe = cat_dum.replace('0', '').groupby('appid', as_index=False).agg(''.join)
ohe_genres = maybe.replace('', '0').astype(int)
ohe_genres.to_csv('../data/processed/export/games_genres_ohe.csv', index=False)
```
